```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('Top10 China-USA in Fortune Global.csv')
```

```python
df2021 = df[df.year == 2021]
```

```python
df2021
```

```python
sns.set(rc={'figure.figsize':(8, 8)})
sns.set_style("white")

pic = sns.scatterplot(data=df2021, x='revenues', y='profits', hue='Country', size='employees', alpha=0.5, palette=['blue','red'], sizes=(100, 2000))

pic.legend(loc='upper right')
pic.set(title = 'Top10 China & USA Companies ranking in Fortune Global 2021',xlabel='2021 Revenue ($M)', ylabel='2021 Profit ($M)')
```

```python
df_avg = df.groupby(by=['year', 'Country']).aggregate('mean')
df_avg= df_avg.round()
df_avg
```

```python
sns.set(rc={'figure.figsize':(8, 8)})
sns.set_style("white")

pic = sns.scatterplot(data=df_avg, x='revenues', y='profits', hue='Country', size='employees', alpha=0.5, palette=['red','blue'], sizes=(100, 500) )
pic_line = sns.lineplot(data=df_avg, x='revenues', y='profits',hue='Country', alpha=0.5, legend = False, palette=['red','blue'])

pic.legend(loc='lower right')
pic.set(title = 'Trend of Top10 China & USA Companies averaging 2019 - 2021',xlabel='Revenue ($M)', ylabel='Profit ($M)')
```
